```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
# 將每篇文章斷詞的list合併成以空格分隔的字串
cut_corpus = []
for i in word_segment:
    cut_corpus.append(' '.join(i))

# 計算詞頻    
text_cv = CountVectorizer()

td_matrix = text_cv.fit_transform(cut_corpus)

print(td_matrix.shape)

#print(text_cv.vocabulary_.keys())
```

```python
import pandas as pd
tfidf = pd.read_csv('./test_data/tdidf.csv')
```

```python
tfidf
```

```python
import re
remove_list=[]
for colname in tfidf.columns:
    if re.match(r'[0-9a-z]',colname):
        remove_list.append(colname)
```

```python
remove_list[-20]
```

```python
# 保留有用的詞
remove_list.remove('5g')
remove_list.remove('iphone')
remove_list.remove('ipad')
#remove_list.remove('vr')
```

```python
len(remove_list)
```

```python
tfidf.drop(columns=remove_list,inplace=True)
```

```python
X = tfidf.values
X.shape
```

```python
from sklearn.decomposition import LatentDirichletAllocation
# n_components代表分成幾個主題
lda = LatentDirichletAllocation(n_components=10,learning_method='batch',n_jobs=-1)
X_topics = lda.fit_transform(X)
```

```python
lda.components_.shape
```

```python
# 印出每個主題前幾重要的關鍵字
# 看起來還是需要刪除停用字
# 或者只是商周不適合分類
n_top_words = 20
feature_names = tfidf.columns

for topic_idx, topic in enumerate(lda.components_):
    print("Topic %d:" % (topic_idx + 1))
    print(" ".join([feature_names[i]
                    for i in topic.argsort()\
                        [:-n_top_words - 1:-1]]))
```
