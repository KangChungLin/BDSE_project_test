```python
import requests
from bs4 import BeautifulSoup
import re
import json
import time
import pandas as pd

def get_urls():
    title_list = []
    url_list = []
    for i in range(1,155):
        try:
            res = requests.get(f'https://wealth.businessweekly.com.tw/BChannelSub.aspx?Channel_no=0004&p={i}')
            soup = BeautifulSoup(res.text,'html.parser')
            a_s = soup.select('div.category_art > h2 > a')
            for a in a_s:
                title_list.append(a.text)
                url_list.append(a.get('href'))
        except:
            print('Error')
            print(f'https://wealth.businessweekly.com.tw/BChannelSub.aspx?Channel_no=0004&p={i}')
        
    return title_list, url_list

def storge(title_list, url_list):
    pd_store= pd.DataFrame(columns = ['Title','URL'])
    pd_store['URL'] = url_list
    pd_store['Title'] = title_list
    
    pd_store.to_csv('businessweekly_urls.csv', index=False)

title_list, url_list = get_urls()
storge(title_list, url_list)
```
